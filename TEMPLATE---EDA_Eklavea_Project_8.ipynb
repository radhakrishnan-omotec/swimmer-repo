1) What is your age group?

````python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
df = pd.read_csv("/content/eda-repo/Output_Enhanced.csv")

# Check the first few rows of the dataframe
print(df.head())

# Check the columns in the dataframe
print(df.columns)

# Assuming "age_group_new" column exists, generate pie chart for its value counts
if 'Age_group_new' in df.columns:
    # Count the occurrences of each value in the 'age_group_new' column
    q1_counts = df['Age_group_new'].value_counts()

    # Plotting the pie chart
    fig, ax = plt.subplots(figsize=(8, 8))
    wedges, labels, autopct = ax.pie(q1_counts, labels=q1_counts.index, autopct='%1.1f%%', startangle=140)

    # Adding legend on top right side
    ax.legend(wedges, q1_counts.index,
              title="Age Groups",
              loc="upper right",
              bbox_to_anchor=(1, 0, 0.5, 1))

    # Adding custom legend labels
    custom_labels = ["0", "1", "2", "3", "4", "5", "6"]
    plt.gca().set_prop_cycle(None)  # Reset color cycle
    for i, custom_label in enumerate(custom_labels):
        plt.plot([], label=custom_label)

    plt.title('Distribution of Age Groups')
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
    plt.show()
else:
    print("Column 'age_group_new' not found in the CSV file.")
````

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file
df = pd.read_csv("/content/eda-repo/Output_Enhanced.csv")

# Check the first few rows of the dataframe
print(df.head())

# Check the columns in the dataframe
print(df.columns)

# Assuming "Q6b_new" (Main Concerns about EVs) and "Q7_new" (Participation in Training Programs) columns exist
if 'Q6b_new' in df.columns and 'Q7_new' in df.columns:
    # Create a contingency table
    contingency_table = pd.crosstab(df['Q6b_new'], df['Q7_new'])

    # Plot heatmap
    plt.figure(figsize=(10, 6))
    sns.heatmap(contingency_table, annot=True, cmap="viridis_r", fmt='g')
    plt.title("Main Concerns about EVs vs. Participation in Training Programs")
    plt.xlabel("Participation in Training Programs")
    plt.ylabel("Main Concerns about EVs")
    plt.tight_layout()
    plt.show()
else:
    print("Columns 'Q6b_new' (Main Concerns about EVs) and/or 'Q7_new' (Participation in Training Programs) not found in the CSV file.")
````
