```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



# Read the CSV file
df = pd.read_csv("/content/swimmer-repo/FINAL-Output_Responses.csv")

# Check the first few rows of the dataframe
print(df.head())

# Check the columns in the dataframe
print(df.columns)

# Assuming "Q1_new" column exists, generate pie chart for its value counts
if 'Q1_new' in df.columns:
    # Count the occurrences of each value in the 'Q1_new' column
    q1_counts = df['Q1_new'].value_counts()

    # Plotting the pie chart
    fig, ax = plt.subplots(1, 2, figsize=(16, 8))



    # Pie chart
    wedges, labels, autopct = ax[0].pie(q1_counts, labels=q1_counts.index, autopct='%1.1f%%', startangle=140)
    ax[0].legend(wedges, q1_counts.index, title="TITLE", loc="upper right", bbox_to_anchor=(1, 0, 0.5, 1))
    custom_labels = [str(i) for i in range(6)]
    plt.gca().set_prop_cycle(None)  # Reset color cycle
    for i, custom_label in enumerate(custom_labels):
        plt.plot([], label=custom_label)
    ax[0].set_title('TITLE')
    ax[0].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.



    # Count plot
    top_four_categories = q1_counts.nlargest(4).index
    sns.countplot(x='Q1_new', data=df[df['Q1_new'].isin(top_four_categories)], ax=ax[1], order=top_four_categories, palette='viridis')
    ax[1].set_title('Count of Top Three TITLE')

    plt.tight_layout()
    plt.show()
else:
    print("Column 'Q1_new' not found in the CSV file.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file
df = pd.read_csv("/content/swimmer-repo/FINAL-Output_Responses.csv")

# Check the first few rows of the dataframe
print(df.head())

# Check the columns in the dataframe
print(df.columns)

# Assuming "Q5_new" and "Gender_new" columns exist
if 'Q5_new' in df.columns and 'Gender_new' in df.columns:
    # Set the figure size
    plt.figure(figsize=(20, 12))

    # Create the countplot
    ax = sns.countplot(x="Q5_new", hue="Gender_new", data=df, palette='viridis')
    plt.title("Gender-based Coping Mechanisms for Pre-Race Anxiety")
    plt.xlabel("anxiety scale")
    plt.ylabel("gender  Count")
    plt.legend(title="anxiety scale")

    # Display count values along with x-axis categories and y-axis counts
    for p in ax.patches:
        height = p.get_height()  # y-axis value (count)
        x_value = p.get_x() + p.get_width() / 2.  # Position for x
        x_category = p.get_x() + p.get_width() / 2.  # Position for x label

        # Extract the x-axis category label
        x_tick_label = ax.get_xticklabels()[int(x_value)].get_text()

        # Annotate with x-axis label, y-axis label, and the respective values
        ax.annotate(f'Likelihood: {x_tick_label}\nCount: {height}', (x_value, height),
                    ha='center', va='center', fontsize=12, color='black', xytext=(0, 10),
                    textcoords='offset points')

    plt.tight_layout()
    plt.show()
else:
    print("Columns 'Q5_new' and/or 'Gender_new' not found in the CSV file.")
```
